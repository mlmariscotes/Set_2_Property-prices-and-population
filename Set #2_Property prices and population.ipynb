```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/FP1.7_Set #2_Property prices and population/population_property_prices.csv')
```

```python
df.head(10)
```


# 1. Create a scatter plot of property prices vs. population.

```python
# Scatter plot of property prices vs. population

plt.figure(figsize=(8, 5))
plt.scatter(df['estimate'], df['avg_price'], color='blue')
plt.title("Scatter Plot of Property Prices vs. Population")
plt.xlabel("Population")
plt.ylabel("Average Property Price ($)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```

# Insight

Here's the scatter plot showing the relationship between population and average property prices for various states in the USA. As depicted, there seems to be a varied relationship between population and property prices, with some states having higher property prices despite a smaller population, and vice versa.



# 2.  Calculate the correlation coefficient between property prices and population. Use the function CORREL() for this task.

```python
# Calculate the correlation coefficient between property prices and population
correlation_coefficient = df['estimate'].corr(df['avg_price'])

correlation_coefficient
```

The correlation coefficient between property prices and population is approximately **0.418**. This indicates a moderate positive linear relationship between the two variables. In other words, as the population increases, there's a tendency for property prices to increase as well, although the relationship isn't particularly strong.

# 3. Create a histogram of the property price distribution for all states.

```python
# Histogram of the property price distribution for all states

plt.figure(figsize=(8, 5))
plt.hist(df['avg_price'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Average Property Prices for All States")
plt.xlabel("Average Property Price ($)")
plt.ylabel("Number of States")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```

# Insight

Here's the histogram showcasing the distribution of average property prices for various states in the USA. The histogram provides insights into how many states fall within certain price ranges.

From the graph, it appears that a significant number of states have property prices in the range of $1,000 to $2,500. There are fewer states with very high or very low average property prices.
